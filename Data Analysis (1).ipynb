### Clean missing values using Random Value Imputation Because This the best way to To maintain distrbuation

```python
def Random_value_imputation(feature):
    random_sample=df[feature].dropna().sample(df[feature].isnull().sum())               
    random_sample.index=df[df[feature].isnull()].index
    df.loc[df[feature].isnull(),feature]=random_sample
```

```python
for col in df:
    Random_value_imputation(col)
```

```python
missing = pd.DataFrame({'Missing Values ' : df.isna().sum().sort_values(ascending=False), 'Percentage of Missing Values ' : (df.isna().sum().sort_values(ascending=False)) / (df.shape[0]) * (100)})
missing
```
